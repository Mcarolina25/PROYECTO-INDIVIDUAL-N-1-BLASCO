```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Cargar el dataset procesado
movies = pd.read_csv("..\Movies\processed_Movies_dataset.csv")

# Informacin general del dataset
print(movies.info())
print(movies.describe())

# Distribucin de pelculas por ao
plt.figure(figsize=(10, 6))
movies['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribucin de pelculas por ao')
plt.xlabel('Ao')
plt.ylabel('Cantidad de pelculas')
plt.show()

# Nube de palabras de los ttulos de las pelculas
text = ' '.join(movies['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de palabras de los ttulos de las pelculas')
plt.show()

# Correlacin entre presupuesto y revenue
plt.figure(figsize=(8, 6))
plt.scatter(movies['budget'], movies['revenue'], alpha=0.5)
plt.title('Relacin entre presupuesto y revenue')
plt.xlabel('Presupuesto')
plt.ylabel('Revenue')
plt.show()
```

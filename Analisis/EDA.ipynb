```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Cargar el dataset procesado
movies = pd.read_csv("..\Movies\processed_Movies_dataset.csv")

# Información general del dataset
print(movies.info())
print(movies.describe())

# Distribución de películas por año
plt.figure(figsize=(10, 6))
movies['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribución de películas por año')
plt.xlabel('Año')
plt.ylabel('Cantidad de películas')
plt.show()

# Nube de palabras de los títulos de las películas
text = ' '.join(movies['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de palabras de los títulos de las películas')
plt.show()

# Correlación entre presupuesto y revenue
plt.figure(figsize=(8, 6))
plt.scatter(movies['budget'], movies['revenue'], alpha=0.5)
plt.title('Relación entre presupuesto y revenue')
plt.xlabel('Presupuesto')
plt.ylabel('Revenue')
plt.show()
```
